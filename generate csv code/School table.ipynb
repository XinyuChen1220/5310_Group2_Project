## School table

```python
import pandas as pd
import numpy as np
import random
import ast 
```

```python
# Load the home listings data
home_listings = pd.read_csv('/Users/chenxinyu/Desktop/home_listings.csv')
```

```python
# Initialize the empty list to store school data extracted from the home listings
schools_list = []
```

```python
# Set school types and levels
# assume  school type to be private or public
# assume school level to be classified into 'Middle', 'High', 'Elementary', 'Pre-K'
school_types = ['Public', 'Private']
school_levels = ['Middle', 'High', 'Elementary', 'Pre-K']
```

```python
# Iterates over each row in the home_listings DataFrame 
# Extracts the city and zip code from the home listing
for idx, row in home_listings.iterrows():
    try:
        schools = ast.literal_eval(row['Nearby Schools'])
    except ValueError:
        continue 

    city = row['City']
    zip_code = row['Zip Code']
    
    for school_dict in schools:
        # Extract school name and rating from 'Nearby Schools' column in home_listings table
        school_name = school_dict['name']
        school_rating = school_dict['rating']
        # Appends a dictionary to schools_list with school information, including a randomly assigned type and level
        schools_list.append({
            'School': school_name,
            'School Rating': school_rating,
            'School Type': random.choice(school_types),
            'School Level': random.choice(school_levels),
            'City': city,
            'ZIP Code': zip_code
        })
```

```python
# Create a DataFrame
schools_df = pd.DataFrame(schools_list)
```

```python
# Save to CSV
schools_df.to_csv('/Users/chenxinyu/Desktop/schools.csv', index=False)
```
